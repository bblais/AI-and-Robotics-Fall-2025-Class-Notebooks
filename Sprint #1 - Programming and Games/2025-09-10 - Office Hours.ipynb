```python
from mplturtle import *
```

```python
def square(size):
    for _ in range(4):
        forward(size)
        right(90)
```

```python
reset()
square(50)
```

```python
reset()

current_color="black"

for j in range(5):
    
    # draw row
    for i in range(5):
        pencolor(current_color)
        square(50)

        if current_color == "black":
            current_color = "red"
        else:
            current_color = "black"
        
        print(i)
        penup()
        forward(80)
        pendown()
    
    # move over
    penup()
    right(90)
    forward(80)
    right(90)
    forward(400)
    right(180)
    pendown()
    
    
```
