```python
from mplturtle import *
```

```python
def square(size):
    for i in range(4):
        forward(size)
        right(90)

def forward(distance,pen="down"):
    from mplturtle import forward as original_forward

    was_up=isup()
    
    if pen == "down":
        pendown()
    else:
        penup()
    
    original_forward(distance)

    if was_up:
        penup()
    else:
        pendown()
```

```python
reset()
square(100)
forward(150,"up")
square(100)
forward(150,"up")
square(100)
forward(150,"up")
```

```python
from random import randint, choice
```

```python
reset()

for i in range(100):
    x,y=randint(-200,200),randint(-200,200)
    size=randint(20,100)
    color=choice(["red","blue","green","black","orange","purple"])
    shape=choice(["square","circle"])
    
    penup()
    goto(x,y)
    pendown()
    
    pencolor(color)
    if shape=="square":
        square(size)
    elif shape=="circle":
        circle(size)
    else:
        raise ValueError("Unknown shape: "+shape)
```
