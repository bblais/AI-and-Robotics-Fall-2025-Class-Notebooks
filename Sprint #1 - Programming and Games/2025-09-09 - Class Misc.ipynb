```python
from mplturtle import *
```

```python
def square(size):
    for i in range(4):
        forward(size)
        right(90)
```

```python
reset()
square(100)
```

```python
def somefunction(size,shape):

    if shape=="square":
        square(size)
```

```python
reset()
somefunction(100,"square")
```

```python
reset()

penup()
goto( (random()*2-1)*100, (random()*2-1)*100 )
pendown()

square(20)

penup()
goto( (random()*2-1)*100, (random()*2-1)*100 )
pendown()

square(20)
```
