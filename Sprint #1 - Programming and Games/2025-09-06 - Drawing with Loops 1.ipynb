```python
def square(size):
    # as if size=#
    for i in range(4):
        forward(size)
        right(90)

def hexagon(size):
    for i in range(6):
        forward(size)
        right(360/6)
```

```python
reset(figsize=(5,5))

size=80

penup()
backward(size//2)
left(90)
forward(size//2)
right(90)
pendown()

square(size)

penup()
forward(size//2)
right(90)
backward(size//2)
left(90)
pendown()


```

```python
def centered_square(size):
    
    penup()
    backward(size//2)
    left(90)
    forward(size//2)
    right(90)
    pendown()
    
    square(size)
    
    penup()
    forward(size//2)
    right(90)
    forward(size//2)
    left(90)
    pendown()
        
```

```python
reset(figsize=(5,5))
centered_square(50)
centered_square(80)
```

```python
reset(figsize=(5,5))

for size in range(10,110,10):
    centered_square(size)
```

```python
def centered_circle(size):
    penup()
    right(90)
    forward(size//2)
    left(90)
    pendown()
    
    circle(size)
    
    penup()
    left(90)
    forward(size//2)
    right(90)
    pendown()
        
```

```python
reset(figsize=(5,5))

for size in range(10,110,10):
    centered_circle(size)
```
