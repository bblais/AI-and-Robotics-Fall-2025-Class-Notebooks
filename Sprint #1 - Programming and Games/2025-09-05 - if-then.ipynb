```python
reset()

for i in range(4):
    forward(101)
    right(90)
```

```python
def square(size):
    for i in range(4):
        forward(size)
        right(90)   

def forward_penup(length):
    penup()
    forward(length)
    pendown()
```

```python
reset()

for i in range(10):
    square(100)
    forward_penup(150)
```

```python
reset()

for bob in range(10):
    print(bob)
    square(100)
    forward_penup(150)
```

```python
reset()

for bob in range(10):

    if bob<4:
        pencolor("green")
    else:
        pencolor("red")
        
    square(100)
    forward_penup(150)
```

```python
def isodd(number):
    return number%2==1

def iseven(number):
    return number%2==0    
```

```python
reset()

for bob in range(10):

    if bob%2==1: # odd
        pencolor("green")
    else:
        pencolor("red")
        
    square(100)
    forward_penup(150)
```

```python
reset()

for bob in range(10):

    if isodd(bob): # odd
        pencolor("green")
    else:
        pencolor("red")
        
    square(100)
    forward_penup(150)
```

```python
reset()

current_color="green"

for bob in range(10):

    if current_color=="green":
        current_color="red"
    else:
        current_color="green"

    pencolor(current_color)
    square(100)
    forward_penup(150)
```
